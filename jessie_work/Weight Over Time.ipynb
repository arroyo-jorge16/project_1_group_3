```python
from pathlib import Path
import pandas as pd
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a file path
filepath = '../jessie_work/summer_olympics.csv'
filepath2 = '../jessie_work/winter_olympics.csv'
# Read in the data. 

summer_df=pd.read_csv(filepath)
winter_df=pd.read_csv(filepath2)

# weight is in KG, height is in CM
```

```python
# Sort the DataFrame by year
summer_sorted = summer_df.sort_values('Year')

# Calculate the average weight for each year
average_weights_summer = summer_sorted.groupby('Year')['Weight'].mean().reset_index()

# Plot the averages by year
plt.figure(figsize=(12, 6))
plt.plot(average_weights_summer['Year'], average_weights_summer['Weight'], marker='o', color='#009F3D', linestyle='--')
plt.title('Average Weight by Year Summer Olympics')
plt.xlabel('Year', style = 'italic')
plt.ylabel('Average Weight (kg)', style = 'italic')
plt.grid(True)
plt.show()
```

```python
# Sort the DataFrame by year
winter_sorted = winter_df.sort_values('Year')

# Calculate the average weight for each year
average_weights_winter = winter_sorted.groupby('Year')['Weight'].mean().reset_index()

# Plot the averages by year
plt.figure(figsize=(12, 6))
plt.plot(average_weights_winter['Year'], average_weights_winter['Weight'], marker='o', color='#009F3D', linestyle='--')
plt.title('Average Weight by Year Summer Olympics')
plt.xlabel('Year', style = 'italic')
plt.ylabel('Average Weight (kg)', style = 'italic')
plt.grid(True)
plt.show()
```

```python
#Calculate and print the median and mean for the 'weight' column
weight_median = summer_df['Weight'].median()
weight_mean = summer_df['Weight'].mean()
weight_max = summer_df['Weight'].max()
weight_min = summer_df['Weight'].min()


print(f"Median Weight: {weight_median}")
print(f"Mean Weight: {weight_mean}")
print(f"Max Weight: {weight_max}")
print(f"Min Weight: {weight_min}")
```

```python
#Calculate and print the median and mean for the 'weight' column
weight_median = winter_df['Weight'].median()
weight_mean = winter_df['Weight'].mean()
weight_max = winter_df['Weight'].max()
weight_min = winter_df['Weight'].min()


print(f"Median Weight: {weight_median}")
print(f"Mean Weight: {weight_mean}")
print(f"Max Weight: {weight_max}")
print(f"Min Weight: {weight_min}")
```

```python
# Plot the averages by year
plt.figure(figsize=(12, 6))

# Plot the Winter Olympics data with a dashed line
plt.plot(average_weights_winter['Year'], average_weights_winter['Weight'], marker='o', color='#0085C7', linestyle='--', label='Winter Olympics')

# Plot the Summer Olympics data with a dashed line and markers
plt.plot(average_weights_summer['Year'], average_weights_summer['Weight'], color='#DF0024', linestyle='--', marker='o', label='Summer Olympics')

plt.title('Average Weight by Year Olympics', fontweight='bold', fontproperties='Arial', size=16)
plt.xlabel('Year', style='italic', size=12, fontproperties='Arial')
plt.ylabel('Average Weight', style='italic', size=12, fontproperties='Arial')
plt.grid(True)
plt.legend()

# Customize the markers on the x-axis to show every four years
plt.xticks(np.arange(average_weights_winter['Year'].min(), average_weights_winter['Year'].max() + 3, 4), rotation=45, fontsize=10, fontname='Arial')

# Set x-axis limits to 2 years beyond the maximum year and 2 years before the minimum year
plt.xlim(average_weights_summer['Year'].min() - 2, average_weights_winter['Year'].max() + 4)


# Save the plot as an image before displaying it
plt.savefig('weight_time_winter_summer.png', dpi=300)

# Display the plot
plt.show()
```
