```python
from pathlib import Path
import pandas as pd
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a file path
filepath = '../jessie_work/summer_olympics.csv'
# Read in the data. 

summer_df=pd.read_csv(filepath)

summer_df.head(10)

# weight is in KG, height is in CM
```

```python
# Sort the DataFrame by year
summer_sorted = summer_df.sort_values('Year')

# Calculate the average height for each year
average_heights_summer = summer_sorted.groupby('Year')['Height'].mean().reset_index()

# Plot the averages by year
plt.figure(figsize=(12, 6))
plt.plot(average_heights_summer['Year'], average_heights_summer['Height'], marker='o', color='#009F3D', linestyle='--')
plt.title('Average Height by Year Summer Olympics')
plt.xlabel('Year', style = 'italic')
plt.ylabel('Average Height (cm)', style = 'italic')
plt.grid(True)
plt.show()
```
