```python
from pathlib import Path
import pandas as pd
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a file path
filepath = '../jessie_work/athlete_events.csv'
# Read in the data. 

og_df=pd.read_csv(filepath)

og_df.head()
```

```python
og_df.info()
```

```python
max_year = og_df["Year"].max()
min_year = og_df["Year"].min()

print("Maximum Year:", max_year)
print("Minimum Year:", min_year)
```

```python
df = og_df[og_df["Year"] >= 1960]
df
```

```python
unique_games_df = df.drop_duplicates(subset='Games')
print(unique_games_df.shape)
unique_games_df.head()
```

```python
# Count the instances of Summer and Winter Olympics after removing duplicates
season_counts = unique_games_df['Season'].value_counts()

print(season_counts)
```

```python
# Sort the filtered DataFrame based on the 'Games' column
sorted_df = df.sort_values(by='Games')

# Display the sorted DataFrame
sorted_df.head(25)
```
